```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "http://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
soup.select('div.chart-wrapper p')
```

```python
song = []
artist = []

num_iter = len(soup.select("p.title-artist"))

for i in range(num_iter):
    song.append(soup.select("p.title-artist")[i].get_text())
    artist.append(soup.select("p.chart-position")[i].get_text())
```

```python
import pandas as pd
```

```python
top100 = pd.DataFrame({"song_title":song,
                       "artist":artist,
                      })
```

```python
song = []
artist = []

num_iter = len(soup.select("div.chart-wrapper p em"))

for i in range(num_iter):
    song.append(soup.select("div.chart-wrapper p cite")[i].get_text())
    artist.append(soup.select("div.chart-wrapper p em")[i].get_text())
```

```python
top100 = pd.DataFrame({"song_title":song,
                       "artist":artist,
                      })
```
